```python
y = df['Label']

X = df[['Source Port','Destination Port', 'Protocol','Flow Duration','ACK Flag Count', 'Fwd Packet Length Std', 'Fwd Packet Length Min','Packet Length Mean','Average Packet Size', 'Min Packet Length','Fwd Packet Length Max']]
```

```python
y.head()
```

```python
seed = 50  # so that the result is reproducible
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.333, random_state = seed)
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier(n_estimators=100)

```

```python

clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)
```

```python
from sklearn import metrics
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, roc_auc_score, roc_curve, f1_score
```

```python
from sklearn.metrics import plot_confusion_matrix
```

```python
plot_confusion_matrix(clf, X_test, y_test)  
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
